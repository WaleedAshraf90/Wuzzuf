```python



# driver.get('https://wuzzuf.net/search/jobs/?q=&a=hpb')
#     # scrape job_titles:
# job_titles= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")            # job_titles منهم واللي هما بالنسبالي دول ال text وده لان التاج ده فيه منه أكتر من واحد بنفس الكلاس ده انا عندي 15 واحد وبعد كده بلوب عليهم وبطلع ال find_element مش find_elements بكلاس معين فعشان كده بنعمل h2 والوظيفه الواحده التاج بتاعها اسمه job_title أنا عندي في الصفحه الواحده 15 وظيفه يعني 15 
# job_titles_lst1= [element.text for element in job_titles]
# job_titles_lst.append(job_titles_lst1)    
#     # scrape company_names:
# company_names= driver.find_elements_by_css_selector("a[class= 'css-17s97q8']")         # find_element مش find_elements بكلاس معين وبما ان التاج ده بنفس كلاسه في منه 15 زيه فبالتالي بعمل a فأنا هنا بشوف الوظايف موجوده في انهي تاج لقيته في التاج اللي اسمه job_titles نفس الكلام زي ال
# company_names_lst1= [element.text.replace('-', '').strip() for element in company_names]   # عشان اشيل المسافات الزياده اللي علي يمين وشمال اسم الشركه بحيث يرجعلي الاسم بس بدون مسافات وزيادات strip وهنا انا بعمل تنضيف لاسم الشركه لان الاسم بيرجعلي في اخره شرطه - فبشلها بأني بستبدلها ب مسافه وبعد كده بعمل company_name واللي هيبقي بالنسبالي ده ال text بعد كده هلوب علي تاج تاج منهم وهطلع منه ال
# company_names_lst.append(company_names_lst1)    
#     # scrape job_location:
# job_location= driver.find_elements_by_css_selector("span[class= 'css-5wys0k']")        # نفس الكلام زي فوق
# job_location_lst1= [element.text for element in job_location]
# job_location_lst.append(job_location_lst1)  
#     # scrape job_posted:
# job_posted_new= driver.find_elements_by_css_selector("div[class= 'css-4c4ojb']")       # نفس الكلام زي فوق
# job_posted_new_lst= [element.text for element in job_posted_new]
# t.sleep(3)
# job_posted_old= driver.find_elements_by_css_selector("div[class= 'css-do6t5g']")       # نفس الكلام زي فوق
# job_posted_old_lst= [element.text for element in job_posted_old]
# t.sleep(3)
# job_posted_all1= [*job_posted_new_lst, *job_posted_old_lst]           # فبكده هيبقوا ليست واحده job_posted_all داخل الليست job_posted_old_lst و job_posted_new_lst واللي معناها فضي محتويات الليست  unbacking ال* يعني
# job_posted_all.append(job_posted_all1)      
#     # scrape job_type:
# job_type= driver.find_elements_by_css_selector("div[class= 'css-1lh32fc']")            # نفس الكلام زي فوق
# job_type_lst1= [element.text.replace('\n', ' - ') for element in job_type]              # ' - ' ب '\n' هنضف بأني هستبدل ال
# job_type_lst.append(job_type_lst1)   
#     # scrape skills:
# job_skills= []
# containers= driver.find_elements_by_css_selector("div[class= 'css-1gatmva e1v1l3u10']")       # كلها skills الكبير اللي شايل ال division ده ال
# for container in containers:       # بتاعت الوظيفه skills المهم هضم العناصر دي كلها مع بعض لانهم كلهم بيرجعوا ال skills اللي بيشيل بردو بعض ال a ثم العنصر span ده فيه اكتر من div ولكن التاني مش الاول الان ال span ثم العنصر Experiances واللي بيشيل ال a ده علي العناصر اللي فيه زي division هلوب علي ال
#     skills= container.find_element_by_css_selector("div[class= 'css-y4udm8']").find_element_by_css_selector("div:nth-child(2)").text
#     skills= skills.replace('\n', '')        # اللي هي علامه النزول لسطر جديد \n من ال text بنضف ال
#     job_skills.append(skills)
    
#     # scrape job_link
# containers= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")     # Job_Skills نفس الكلام زي ال
# links= []
# for container in containers:
#     job_link= container.find_element_by_css_selector("a[class= 'css-o171kl']").get_attribute('href')
#     links.append(job_link)
#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
job_titles_lst=[]
company_names_lst=[]
job_location_lst=[]
job_posted_all=[]
job_skills= []
job_type_lst=[]
links= []
for i in range(1,601):
    
    driver.get(f'https://wuzzuf.net/search/jobs/?a=hpb&q=&start={i}')
    
     # scrape job_titles:
    job_titles= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")            # job_titles منهم واللي هما بالنسبالي دول ال text وده لان التاج ده فيه منه أكتر من واحد بنفس الكلاس ده انا عندي 15 واحد وبعد كده بلوب عليهم وبطلع ال find_element مش find_elements بكلاس معين فعشان كده بنعمل h2 والوظيفه الواحده التاج بتاعها اسمه job_title أنا عندي في الصفحه الواحده 15 وظيفه يعني 15 
    job_titles_lst1= [element.text for element in job_titles]
    job_titles_lst.extend(job_titles_lst1) 
    # scrape company_names:
    company_names= driver.find_elements_by_css_selector("a[class= 'css-17s97q8']")         # find_element مش find_elements بكلاس معين وبما ان التاج ده بنفس كلاسه في منه 15 زيه فبالتالي بعمل a فأنا هنا بشوف الوظايف موجوده في انهي تاج لقيته في التاج اللي اسمه job_titles نفس الكلام زي ال
    company_names_lst1= [element.text.replace('-', '').strip() for element in company_names]   # عشان اشيل المسافات الزياده اللي علي يمين وشمال اسم الشركه بحيث يرجعلي الاسم بس بدون مسافات وزيادات strip وهنا انا بعمل تنضيف لاسم الشركه لان الاسم بيرجعلي في اخره شرطه - فبشلها بأني بستبدلها ب مسافه وبعد كده بعمل company_name واللي هيبقي بالنسبالي ده ال text بعد كده هلوب علي تاج تاج منهم وهطلع منه ال
    company_names_lst.extend(company_names_lst1) 
    # scrape job_location:
    job_location= driver.find_elements_by_css_selector("span[class= 'css-5wys0k']")        # نفس الكلام زي فوق
    job_location_lst1= [element.text for element in job_location]
    job_location_lst.extend(job_location_lst1)  
    # scrape job_posted:
    job_posted_new= driver.find_elements_by_css_selector("div[class= 'css-4c4ojb']")       # نفس الكلام زي فوق
    job_posted_new_lst= [element.text for element in job_posted_new]
    t.sleep(3)
    job_posted_old= driver.find_elements_by_css_selector("div[class= 'css-do6t5g']")       # نفس الكلام زي فوق
    job_posted_old_lst= [element.text for element in job_posted_old]
    t.sleep(3)
    job_posted_all1= [*job_posted_new_lst, *job_posted_old_lst]           # فبكده هيبقوا ليست واحده job_posted_all داخل الليست job_posted_old_lst و job_posted_new_lst واللي معناها فضي محتويات الليست  unbacking ال* يعني
    job_posted_all.extend(job_posted_all1)   
    # scrape job_type:
    job_type= driver.find_elements_by_css_selector("div[class= 'css-1lh32fc']")            # نفس الكلام زي فوق
    job_type_lst1= [element.text.replace('\n', ' - ') for element in job_type]              # ' - ' ب '\n' هنضف بأني هستبدل ال
    job_type_lst.extend(job_type_lst1)  
    # scrape skills:
    
    containers= driver.find_elements_by_css_selector("div[class= 'css-1gatmva e1v1l3u10']")       # كلها skills الكبير اللي شايل ال division ده ال
    for container in containers:       # بتاعت الوظيفه skills المهم هضم العناصر دي كلها مع بعض لانهم كلهم بيرجعوا ال skills اللي بيشيل بردو بعض ال a ثم العنصر span ده فيه اكتر من div ولكن التاني مش الاول الان ال span ثم العنصر Experiances واللي بيشيل ال a ده علي العناصر اللي فيه زي division هلوب علي ال
        skills= container.find_element_by_css_selector("div[class= 'css-y4udm8']").find_element_by_css_selector("div:nth-child(2)").text
        skills= skills.replace('\n', '')        # اللي هي علامه النزول لسطر جديد \n من ال text بنضف ال
        job_skills.append(skills)
    
    # scrape job_link
    containers= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")     # Job_Skills نفس الكلام زي ال
    links= []
    for container in containers:
        job_link= container.find_element_by_css_selector("a[class= 'css-o171kl']").get_attribute('href')
        links.append(job_link)
      
     # Save all data in csv files:
    from itertools import zip_longest
    from csv import writer

    file_list= [job_titles_lst, company_names_lst, job_location_lst, job_posted_all, job_type_lst, job_skills, links]
    exported= zip_longest(*file_list)

    with open(f' waleed.csv', 'w', encoding="utf-8") as csv_file:
        wr= writer(csv_file)
        wr.writerow(['Job Title', 'Company Name', 'Job Location', 'Job Posted', 'Jop Type', 'Job Skill', 'Job Link'])
        wr.writerows(exported) 
```

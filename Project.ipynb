# ML

```python
df
```

```python
df.drop(['Job Posted','Jop Type','Time','Month Or Hours','Job Location','Company Name'],axis=1,inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Job Title',   'Job Skill',
       ], drop_first=True)
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = scaler.fit_transform(df)
```

```python
model = KMeans(n_clusters=3)
clusters = model.fit_predict(df_scaled)
clusters
```

```python
df['Clusters'] = clusters
df
```

```python
scores = []
for i in range(1, 50):
    model = KMeans(n_clusters=i)
    model.fit(df_scaled)
    scores.append(model.inertia_)
plt.plot(range(1, 50), scores)
plt.title("Elbow Method")
plt.xlabel("num of clusters")
plt.ylabel("Score")
```

```python
# from the scores, turns out the best clusters number is 5
model = KMeans(n_clusters=30)
clusters = model.fit_predict(df_scaled)
clusters
```

```python
df['Clusters'] = clusters
df
```
